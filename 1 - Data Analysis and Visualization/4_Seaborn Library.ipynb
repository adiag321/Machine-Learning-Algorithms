# Seaborn Library Part 1

Importing the Libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
insur = pd.read_csv("E:\Jupyter Notebook\Dataset/insurance.csv")
insur.head()
```

```python
# setting a theme for the plots
sns.set_theme()
```

### Relplot

```python
sns.relplot(data=insur, x ='age',y='bmi',col='region',hue='smoker',style='smoker',size='children')
```

### Scatter Plot

```python
plt.figure(figsize=(20,10))
sns.scatterplot(x='age',y='charges',hue='region',data=insur)
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.scatterplot(x='bmi',y='charges',hue='region',data=insur)
plt.show()
```

### Count Plot

```python
sns.countplot(x='region',data=insur)
```

```python
sns.countplot(x='smoker',data=insur)
```

### Bar Chart

```python
sns.barplot(x='region',y='age',hue='smoker',data=insur)
```

```python
sns.barplot(x='smoker',y='age',hue='region',data=insur)
```

### Distribution Plot

```python
sns.distplot(insur['bmi'])
```

Correlation:
- Positive Correlation
- Negative Correlation

### Heat Map

```python
correlation = insur.corr()
```

```python
# constructing a Heat Map
plt.figure(figsize=(10,10))
sns.heatmap(data = correlation, cbar=True, fmt='.1f', annot=True, annot_kws={'size':9}, cmap='Blues')
```

# Seaborn Library Part 2

Dataset: 

```python
toyo = pd.read_csv("E:\Jupyter Notebook\Dataset/ToyotaCorolla.csv")
toyo.sample(10)
```

### SCATTER PLOT - 2 VARIABLE [REGPLOT]

```python
plt.style.use('ggplot')
plt.figure(figsize=(20,8))
sns.regplot(x = toyo['Age'], y = toyo["Price"])
plt.show()
```

### SCATTER PLOT - 3 VARIABLE [LMPLOT]

```python
sns.lmplot(x = "Age", y = "Price", data = toyo, fit_reg = False, hue = "FuelType", legend = True, height = 8)
```

### Histogram

```python
plt.figure(figsize = (20,10))
sns.distplot(toyo['Age'], kde = False)
plt.show()
```

### Bar Plot:

```python
sns.countplot(x = "MetColor", data = toyo)
```

```python
plt.figure(figsize = (20,10))
sns.countplot(x = "Age", data = toyo)
plt.show()
```

### Grouped Bar Plot:

```python
sns.countplot(x = "FuelType", data = toyo, hue = "Automatic")
```

### Box and Whiskers Plot:

```python
plt.figure(figsize=(8,6))
sns.boxplot(y=toyo["Price"])
plt.show()
```
