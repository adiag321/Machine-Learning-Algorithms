# SEABORN ESSENTIALS

```python
import seaborn as sns
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
import matplotlib.pyplot as plt
```

## Built-in Datasets

```python
sns.get_dataset_names()
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
# loading additional datasets  
tips = sns.load_dataset("tips")  
mpg = sns.load_dataset("mpg")  
fmri = sns.load_dataset("fmri")  
car_crashes = sns.load_dataset("car_crashes")
```

### FOR DISTRIBUTION PLOTS WE USE -

By distribution I mean few things: frequency distribution, probability distribution or just spread of the data with respect to central values (mean, median etc).

1. Histogram plot
2. Kernel Density plot
3. Box plot

```python
# Bar histogram  
sns.distplot(mpg["mpg"], kde = False)
```

```python
# Line histogram/kernel density plot  
sns.kdeplot(mpg["mpg"], shade = True)
```

```python
# boxplot  
sns.boxplot(x = mpg["mpg"])
```

### FOR RELATIONSHIPS PLOTS WE USE - 
Kids grow taller with age — that’s a relationship between two variables: height and age.
The bigger the house, the higher the price — that’s another kind of relationship between the variables: floor size and price.
Relational plots display how variables are associated with each other.

1. Scatter plot
2. Trendline/Linear model
3. Bar chart
4. Line chart
5. Stacked Bar chart
6. Stacked Area

```python
# scatterplot  
sns.scatterplot(x = "sepal_length", y = "petal_length", data = iris)
```

```python
# trend line  
sns.lmplot(x = "sepal_length", y = "petal_length", data = iris, scatter = False)
```

```python
# bar chart  
sns.barplot(x = "day", y = "total_bill", data = tips)
```

```python
# line chart/time series  
sns.lineplot(x="timepoint", y="signal", hue="event", data=fmri)
```

```python
# Stacked Bar Chart
# Initialize the matplotlib figure  
f, ax = plt.subplots(figsize=(10, 15))# Plot the total crashes  
sns.set_color_codes("pastel")  
sns.barplot(x="total", y="abbrev", data=car_crashes,label="Total", color="b") # Plot the crashes where alcohol was involved  
sns.set_color_codes("muted")  
sns.barplot(x="alcohol", y="abbrev", data=car_crashes, label="Alcohol-involved", color="b")# Add a legend and informative axis label  
ax.legend(ncol=2, loc="upper right", frameon=True)
```

```python
# stacked area  
import numpy as np  
import matplotlib.pyplot as plt # Data  
x = range(1,6)  
y = [[2,4,5,6,8], [2,3,6,7,10], [2,6,7,8,5]]# Plot  
plt.stackplot(x,y, labels=['X','Y','Z'])  
plt.legend(loc='upper left')  
plt.show()
```
