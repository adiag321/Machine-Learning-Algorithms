```python
import pandas as pd

df=pd.read_csv('AB_NYC_2019.csv')
df.head(10)
```

```python
df['price'].describe()
```

```python
df.shape
```

```python
min_thresh,max_thresh=df['price'].quantile([0.008,0.99])
min_thresh,max_thresh
```

```python
df[df['price']<min_thresh]
```

```python
df[df['price']>max_thresh]
```

```python
df2=df[(df['price']<max_thresh) & (df['price']>min_thresh)]
```

```python
df.shape[0]-df2.shape[0]   # So, we removed 896 outlier
```
