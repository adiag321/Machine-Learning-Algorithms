```python
import pandas as pd

df=pd.read_csv('./Dataset/heights.csv')
df.head()
```

```python
df.shape
```

## Detecting Outlier

```python
max_threshold=df['height'].quantile(0.95)   # Above 95% quantile
max_threshold
```

```python
df[df['height']>max_threshold] # It's obviously outlier
```

```python
min_threshold=df['height'].quantile(0.05)     ## Below 5% quantile
min_threshold
```

```python
df[df['height']<min_threshold] # it's also outlier
```

## Removing Outlier

```python
df[(df['height']<max_threshold) & (df['height']>min_threshold)]  # Removing Outlier
```

```python
df=pd.read_csv('bhp.csv')
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
min_thresh,max_thresh=df["price_per_sqft"].quantile([0.01,0.9999]) # passing array as quantile argument
min_thresh,max_thresh
```

```python
df[df["price_per_sqft"]<min_thresh] # below minimum threshold looks like outlier
```

```python
df[df["price_per_sqft"]>max_thresh] # above maximum threshold also looks like outlier
```

```python
df2=df[(df["price_per_sqft"]<max_thresh) & (df["price_per_sqft"]>min_thresh)] #Removing outlier
```

```python
df.shape[0]-df2.shape[0]  # So, we removed 44 outlier so far
```
