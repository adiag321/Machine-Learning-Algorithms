# Crosstab

### N.B. : Use 'Shift+Tab' to find documentation of the function

```python
import pandas as pd
df = pd.read_excel("survey.xls")
df
```

```python
pd.crosstab(df.Nationality,df.Handedness) # first parameter indicate what we want in x-axis
```

```python
pd.crosstab(df.Sex,df.Handedness)
```

```python
pd.crosstab(df.Sex,df.Handedness, margins=True)  # margin parameter add 'All' column
```

```python
pd.crosstab(df.Sex, [df.Handedness,df.Nationality], margins=True) # Multi-index column and rows
```

```python
pd.crosstab([df.Nationality, df.Sex], [df.Handedness], margins=True)
```

```python
pd.crosstab(df.Sex, df.Handedness, normalize='index') # Normalize(Divide all by same value)
```

```python
import numpy as np
pd.crosstab(df.Sex, df.Handedness, values=df.Age, aggfunc=np.average) # Aggfunc and values
```
