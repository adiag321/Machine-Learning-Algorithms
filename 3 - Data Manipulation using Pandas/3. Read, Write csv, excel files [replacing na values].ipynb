# Read, write  csv,excel files

```python
import pandas as pd

df = pd.read_csv("stock.csv")
df
```

```python
df = pd.read_csv("stock.csv", header=1)
df = pd.read_csv("stock.csv", skiprows=1) #skiprows & header kind of same
df
```

```python
df = pd.read_csv("stock.csv", header=None, names = ["ticker","eps","revenue","price","people"])
                                                  #for missing header have to use 'header=None' and define column names
df
```

```python
df = pd.read_csv("stock.csv",  nrows=2)           #using specific number of rows
df
```

```python
df = pd.read_csv("stock.csv", na_values=["n.a.", "not available"]) #values('n.a','not available') replacing with na_values(NaN)
df
```

```python
df = pd.read_csv("stock.csv",  na_values={                #replacing with NaN using dictionary
        'eps': ['not available'],
        'revenue': [-1],
        'people': ['not available','n.a.'],
        'price': ['n.a.']
    })
df
```

```python
df.to_csv("new.csv", columns=["tickers","price"], index=False) #saving csv file using specific column and no indexing
```

```python
def convert_people_cell(cell):
    if cell=="n.a.":
        return 'Sam Walton'
    return cell

def convert_price_cell(cell):
    if cell=="n.a.":
        return 50
    return cell
    
df = pd.read_excel("stock_data.xlsx","Sheet1", converters= {        #Reading csv file and using converter
        'people': convert_people_cell,
        'price': convert_price_cell
    })
df
```

```python
df.to_excel("new.xlsx", sheet_name="stocks", index=False, startrow=2, startcol=1) #saving excel defining starting row & starting column
```

### Write two dataframes to two separate sheets in excel

```python
df_stocks = pd.DataFrame({
    'tickers': ['GOOGL', 'WMT', 'MSFT'],
    'price': [845, 65, 64 ],
    'pe': [30.37, 14.26, 30.97],
    'eps': [27.82, 4.61, 2.12]
})

df_weather =  pd.DataFrame({
    'day': ['1/1/2017','1/2/2017','1/3/2017'],
    'temperature': [32,35,28],
    'event': ['Rain', 'Sunny', 'Snow']
})
```

```python
with pd.ExcelWriter('stocks_weather.xlsx') as writer:   #using pd.ExcelWriter()
    df_stocks.to_excel(writer, sheet_name="stocks")
    df_weather.to_excel(writer, sheet_name="weather")
```
