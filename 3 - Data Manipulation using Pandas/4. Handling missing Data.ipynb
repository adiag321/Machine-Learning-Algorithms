# Handling missing Data

### 1. <font color='red'>Fill na Values</font>

```python
import pandas as pd
df = pd.read_csv("weather_data.csv",parse_dates=['day'])
print(type(df.day[0])) #parse-date actually convert day column to date types
df
```

```python
new_df = df.fillna(0)                      #Fill na values
new_df
```

```python
new_df = df.fillna({                  #Fill na values using dictionary
        'temperature': 0,
        'windspeed': 0,
        'event': 'No Event'
    })
new_df
```

```python
new_df = df.fillna(method="ffill")         #fillna using method='ffill' (forward fill)
new_df
```

```python
new_df = df.fillna(method="bfill", axis="columns") # axis is either "index" or "columns"
new_df                                             #this time fillna using backward fill using row
```

```python
new_df = df.fillna(method="ffill",limit=1) #using limit in fillna function
new_df
```

### 2. <font color='red'>Interpolate</font>

```python
new_df = df.interpolate()         #using interpolate 
new_df                            #by default use linear interpolation
```

```python
df.set_index('day',inplace=True)
new_df = df.interpolate(method="time")  #interpolate using time method
new_df                                  #time method making better guess #considering dates for temperatures 32>>29>>28
```

### 3. <font color='red'>Drop na Values</font>

```python
new_df = df.dropna()
new_df                           #Droping every column which have at least one na value
```

```python
new_df = df.dropna(how='all')     #Droping every column which row have na values for all
new_df                            # Using "how" parameters in dropna
```

```python
new_df = df.dropna(thresh=2)     #using thresh parameter in dropna
new_df                           #Droping every column which have at least two na values
```

```python
dt = pd.date_range("01-01-2017","01-11-2017")                 # If We want to use every missing dates
idx = pd.DatetimeIndex(dt)      #Using pandas DatetimeIndex
df.reindex(idx)                 #reindexing
```

### 4. <font color='red'>Data Replace</font>

```python
import numpy as np
df = pd.read_csv("weather_data2.csv")
df
```

```python
new_df = df.replace(-99999, value=np.NaN)  #replacing -99999 with NaN values
new_df                                     #np.NaN
```

```python
new_df = df.replace(to_replace=[-99999,-88888], value=0) #using to_replace passing list for replacing
new_df
```

```python
new_df = df.replace({                      #passing dictionary  # replacing specific column values
        'temperature': -99999,
        'windspeed': -99999,
        'event': '0'
    }, np.nan)    

new_df
```

```python
new_df = df.replace({                     # Replacing by mapping
        -99999: np.nan,                   # change to specific values for specific values
        'no event': 'Sunny',
    })
new_df
```

```python
# when windspeed is 6 mph, 7 mph etc. & temperature is 32 F, 28 F etc.
new_df = df.replace({'temperature': '[A-Za-z]', 'windspeed': '[a-z]'},'', regex=True) # Using regular expression 
new_df                                  # Removing a,b,c,d from numbers
```
