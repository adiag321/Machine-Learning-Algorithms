# Merge Dataframe

```python
import pandas as pd
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75],
})
df2
```

```python
df3 = pd.merge(df1, df2, on="city")             #  Merging two dataframe,on city column
df3
```

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35, 38],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "humidity": [65,68,71],
})
df2
```

```python
df3=pd.merge(df1,df2,on="city",how="inner")    # By default, how=inner join like A^B in higher mathematics
df3
```

```python
df3=pd.merge(df1,df2,on="city",how="outer")
df3
```

```python
df3=pd.merge(df1,df2,on="city",how="left")
df3
```

```python
df3=pd.merge(df1,df2,on="city",how="right")
df3
```

```python
df3=pd.merge(df1,df2,on="city",how="outer",indicator=True) # Indicator= which data from which dataframe
df3
```

```python
df3= pd.merge(df1,df2,on="city",how="outer", suffixes=('_first','_second'))
df3
```

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35,38],
    "humidity": [65,68,71, 75]
})
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "temperature": [21,14,35],
    "humidity": [65,68,71]
})

df3= pd.merge(df1,df2,on="city",how="outer", suffixes=('_first','_second')) # Using suffixes for indicating
df3
```
