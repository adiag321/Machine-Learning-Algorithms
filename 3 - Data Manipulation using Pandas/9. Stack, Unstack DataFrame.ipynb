# Reshape dataframe using stack/unstack

```python
import pandas as pd
df = pd.read_excel("stocks.xlsx",header=[0,1])
df
```

```python
df.stack()       #by default stack takes most inner level and transpose it
```

```python
df.stack(level=0) #specific level transpose
```

```python
df_stacked=df.stack()
df_stacked
```

```python
df_stacked.unstack()            # Unstack
```

```python
df2 = pd.read_excel("stocks_3_levels.xlsx",header=[0,1,2]) # 3 levels of column headers
df2
```

```python
df2.stack()
```

```python
df2.stack(level=2)
```
