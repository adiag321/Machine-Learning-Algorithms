```python
# Importing all the files

import pandas as pd
import numpy as np
import sklearn.datasets as sd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
# Evalution
from sklearn.metrics import accuracy_score
import sklearn.metrics

import warnings
warnings.filterwarnings("ignore")
```

```python
# Loading the dataset
data = sd.load_iris()
display(data)
# Converting the data into Pandas Dataframe
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target # Adding target variable
df.head()
```

```python
df.shape
```

```python
# Splitting target and predictor variable
X = df.iloc[:,:-1]
y = df.target
print("Predictor variable - ", X)
print(X.shape)
print("Target variable - ", y)
print(y.shape)
```

```python
# Splitting the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=42)
```

```python
# Function for evaluation

def evaluate(y_train, predict_train, y_test, predict_test):
  ## Accuracy Metric
  # Accuray Score on train dataset
  accuracy_train = accuracy_score(y_train, predict_train)
  print('accuracy_score on train dataset : ', round((accuracy_train)*100,3), '%')

  # Accuracy Score on test dataset
  accuracy_test = accuracy_score(y_test, predict_test)
  print('accuracy_score on test dataset : ', round((accuracy_test)*100,3), '%')
```

## 1. Logistic Regression

```python
reg = LogisticRegression()
reg.fit(X_train, y_train)
# Predict train class
predict_train = reg.predict(X_train)
# Predict test class 
predict_test = reg.predict(X_test)
```

```python
evaluate(y_train, predict_train, y_test, predict_test)
```

```python
reg = LogisticRegression()
reg.fit(X_train, y_train)
y_pred = reg.predict(X_test)
print(accuracy_score(y_test, y_pred))
```

```python
reg.score(X,y)
```

## 2. Decision Trees

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(X_train, y_train)
y_pred = dtc.predict(X_test)
print(accuracy_score(y_test, y_pred))
```
