## <a href = 'https://pub.towardsai.net/difference-between-normalization-and-standardization-745030eaf96f'>Normalization V/s Standardizaton</a>


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import MaxAbsScaler
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import StandardScaler
sns.set(font_scale=1.5)
```

```python
df = sns.load_dataset('tips')
df = df[['total_bill', 'tip']]
df.head()
```

```python
### Plot function
def original_plot(data, col1, col2):
    fig, axes = plt.subplots(nrows = 1, ncols = 3, figsize=(18,5))
    sns.histplot(data = data, x = col1, ax = axes[0])
    sns.histplot(data = data, x = col2, ax = axes[1])
    sns.scatterplot(data = data, x = col1, y = col2, ax = axes[2])
    plt.show()
```

```python
original_plot(data = df, col1 = 'total_bill', col2 = 'tip')
```

### 1. Normalization - A) MinMax Scalar

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
minmaxscaler_df = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
```

```python
minmaxscaler_df
```

```python
### plot for showing the difference in the plots
def diff_plot(data, mod_data, col1, col2):
    fig, axes = plt.subplots(nrows = 2, ncols = 3, figsize=(18,5))
    sns.histplot(data = data, x = col1, ax = axes[0,0])
    sns.histplot(data = data, x = col2, ax = axes[0,1])
    sns.scatterplot(data = data, x = col1, y = col2, ax = axes[0,2])
    
    sns.histplot(data = mod_data, x = col1, ax = axes[1,0])
    sns.histplot(data = mod_data, x = col2, ax = axes[1,1])
    sns.scatterplot(data = mod_data, x = col1, y = col2, ax = axes[1,2])
    plt.show()
```

```python
diff_plot(data = df, mod_data = minmaxscaler_df, col1 = 'total_bill', col2 = 'tip')
```

### 1. Normalization - B) MinAbScaler

```python
from sklearn.preprocessing import MaxAbsScaler
scaler = MaxAbsScaler()
maxabsscaler_df = pd.DataFrame(scaler.fit_transform(df), columns = df.columns)
```

```python
diff_plot(data = df, mod_data = maxabsscaler_df, col1 = 'total_bill', col2 = 'tip')
```

### 1. Normalization - C) RobustScaler

```python
from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()
robustscaler_df = pd.DataFrame(scaler.fit_transform(df), columns = df.columns)
robustscaler_df.head()
```

```python
diff_plot(data = df, mod_data = robustscaler_df, col1 = 'total_bill', col2 = 'tip')
```

### 2. Stamdardization

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
standardscaler_df = pd.DataFrame(scaler.fit_transform(df), columns = df.columns)
standardscaler_df.head()
```

```python
diff_plot(data = df, mod_data = standardscaler_df, col1 = 'total_bill', col2 = 'tip')
```
